### Example 5. Fill Dummy Values into Missing Values.

```python
diabetes = pd.read_csv('pima-indians-diabetes data.csv', index_col=0)
```

```python
from numpy.random import rand
def fill_dummy_values(df, scaling_factor=.075):
    df_dummy = df.copy(deep=True)
    for col_name in df_dummy:
        col = df_dummy[col_name]
        col_null = col.isnull()    
        # Calculate number of missing values in column 
        num_nulls = col_null.sum()
        # Calculate column range
        col_range = col.max() - col.min()
        # Scale the random values to scaling_factor times col_range
        dummy_values = (rand(num_nulls) - 2) * scaling_factor * col_range + col.min()
        col[col_null] = dummy_values
    return df_dummy
```

```python
# Fill dummy values in diabetes_dummy
diabetes_dummy = fill_dummy_values(diabetes)

# Sum the nullity of Skin_Fold and BMI
nullity = diabetes.Skin_Fold.isnull() | diabetes.BMI.isnull()

# Create a scatter plot of Skin Fold and BMI 
diabetes_dummy.plot(x='Skin_Fold', y='BMI', kind='scatter', alpha=0.5, 
                    
                    # Set color to nullity of BMI and Skin_Fold
                    c=nullity, 
                    cmap='rainbow', 
                    figsize=(12, 10));
```

```python
# Fill dummy values in diabetes_dummy
diabetes_dummy = fill_dummy_values(diabetes)

# Sum the nullity of Skin_Fold and BMI
nullity = diabetes.Skin_Fold.isnull() | diabetes.Serum_Insulin.isnull()

# Create a scatter plot of Skin Fold and BMI 
diabetes_dummy.plot(x='Skin_Fold', y='Serum_Insulin', kind='scatter', alpha=0.5, 
                    
                    # Set color to nullity of BMI and Skin_Fold
                    c=nullity, 
                    cmap='rainbow', 
                    figsize=(10, 8));
```

### Example 1. Mean, Median, Mode and Constant Value Imputation.

```python
from sklearn.impute import SimpleImputer
```

```python
diabetes = pd.read_csv('pima-indians-diabetes data.csv', index_col=0)
```

#### Impute with mean

```python
# Make a copy of diabetes
diabetes_mean = diabetes.copy(deep=True)

# Create mean imputer object
mean_imputer = SimpleImputer(strategy='mean')

# Impute mean values in the DataFrame diabetes_mean
diabetes_mean.iloc[:, :] = mean_imputer.fit_transform(diabetes_mean)
```

#### Impute with median

```python
# Make a copy of diabetes
diabetes_median = diabetes.copy(deep=True)

# Create median imputer object
median_imputer = SimpleImputer(strategy='median')

# Impute median values in the DataFrame diabetes_median
diabetes_median.iloc[:, :] = median_imputer.fit_transform(diabetes_median)
```

#### Impute with mode

```python
# Make a copy of diabetes
diabetes_mode = diabetes.copy(deep=True)

# Create mode imputer object
mode_imputer = SimpleImputer(strategy='most_frequent')

# Impute using most frequent value in the DataFrame mode_imputer
diabetes_mode.iloc[:, :] = mode_imputer.fit_transform(diabetes_mode)
```

#### Impute with constant value

```python
# Make a copy of diabetes
diabetes_constant = diabetes.copy(deep=True)

# Create median imputer object
constant_imputer = SimpleImputer(strategy='constant', fill_value=0)

# Impute missing values to 0 in diabetes_constant
diabetes_constant.iloc[:, :] = constant_imputer.fit_transform(diabetes_constant)
```

#### Visualize the Imputations

```python
# Set nrows and ncols to 2
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
nullity = diabetes.Serum_Insulin.isnull() | diabetes.Glucose.isnull()

# Create a dictionary of imputations
imputations = {'Mean Imputation': diabetes_mean, 'Median Imputation': diabetes_median, 
               'Most Frequent Imputation': diabetes_mode, 'Constant Imputation': diabetes_constant}

# Loop over flattened axes and imputations
for ax, df_key in zip(axes.flatten(), imputations):
    # Select and also set the title for a DataFrame
    imputations[df_key].plot(x='Serum_Insulin', y='Glucose', kind='scatter', 
                          alpha=0.5, c=nullity, cmap='rainbow', ax=ax, 
                          colorbar=False, title=df_key);
```

___Notice:___ These imputations are portrayed as a straight line and don't adjust to the shape of the data.

### Example 2. Filling Missing Time-Series Data with `ffill` and `bfill`

```python
airquality = pd.read_csv('air-quality.csv')
airquality
```

```python
msno.matrix(airquality);
```

#### Forard-Fill

```python
airquality_ffill = airquality.copy()
# Print prior to imputing missing values
print(airquality_ffill[30:40])

# Fill NaNs using forward fill
airquality_ffill.fillna(method='ffill', inplace=True)

# Print after imputing missing values
print()
print(airquality_ffill[30:40])
```

#### Backward-Fill

```python
airquality_bfill = airquality.copy()
# Print prior to imputing missing values
print(airquality_bfill[30:40])

# Fill NaNs using backward fill
airquality_bfill.fillna(method='bfill', inplace=True)

# Print after imputing missing values
print()
print(airquality_bfill[30:40])
```

### Example 3. Interpolate Missing Time-Series Data with `linear` ,  `quadratic`, `nearest` Methods

```python
airquality = pd.read_csv('air-quality.csv')
airquality
```

#### Linear Filling

```python
airquality_linear = airquality.copy()
# Print prior to interpolation
print(airquality_linear[30:40])

# Interpolate the NaNs linearly
airquality_linear.interpolate(method='linear', inplace=True)

# Print after interpolation
print(airquality_linear[30:40])
```

#### Quadratic Filling

```python
airquality_quadratic = airquality.copy()
# Print prior to interpolation
print(airquality_quadratic[30:40])

# Interpolate the NaNs linearly
airquality_quadratic.interpolate(method='linear', inplace=True)

# Print after interpolation
print(airquality_quadratic[30:40])
```

#### Nearest Filling

```python
airquality_nearest = airquality.copy()
# Print prior to interpolation
print(airquality_nearest[30:40])

# Interpolate the NaNs linearly
airquality_nearest.interpolate(method='linear', inplace=True)

# Print after interpolation
print(airquality_nearest[30:40])
```

### Example 4. Visualize Different Imputation Methods

#### Forward Imputation Visualization

```python
# Impute airquality DataFrame with ffill method
ffill_imputed = airquality.fillna(method='ffill')

# Plot the imputed DataFrame ffill_imp in red dotted style 
ffill_imputed['Ozone'].plot(color='red', marker='o', linestyle='dotted', figsize=(30, 5))

# Plot the airquality DataFrame with title
airquality['Ozone'].plot(title='Ozone', marker='o', figsize=(30, 5));
```

#### Backward Imputation Visualization

```python
# Impute airquality DataFrame with bfill method
bfill_imputed = airquality.fillna(method='bfill')

# Plot the imputed DataFrame bfill_imp in red dotted style 
bfill_imputed['Ozone'].plot(color='red', marker='o', linestyle='dotted', figsize=(30, 5))

# Plot the airquality DataFrame with title
airquality['Ozone'].plot(title='Ozone', marker='o', figsize=(30, 5))
```

#### `linear`, `quadratic`, `nearest` Interpolations Visualization

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(30, 20))

interpolation_linear = airquality.interpolate(method='linear')

interpolation_quadratic = airquality.interpolate(method='quadratic')

interpolation_nearest = airquality.interpolate(method='nearest')

interpolations = {'Linear Interpolation': interpolation_linear, 'Quadratic Interpolation': interpolation_quadratic, 
                  'Nearest Interpolation': interpolation_nearest}

for ax, df_key in zip(axes.flatten(), interpolations):
    interpolations[df_key].Ozone.plot(color='red', marker='o', linestyle='dotted', ax=ax)
    airquality.Ozone.plot(marker='o', ax=ax)
    ax.set_title(label=df_key + '- Ozone', fontdict={"fontsize": 18});
```

#### Combined Together: Imputation Technique Comparisons

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(30, 20))

# Impute airquality DataFrame with ffill method
ffill_imputed = airquality.fillna(method='ffill')

# Impute airquality DataFrame with bfill method
bfill_imputed = airquality.fillna(method='bfill')

# Interpolate airquality DataFrame with `linear` method
interpolation_linear = airquality.interpolate(method='linear')

# Interpolate airquality DataFrame with `quadratic` method
interpolation_quadratic = airquality.interpolate(method='quadratic')

# Interpolate airquality DataFrame with `nearest` method
interpolation_nearest = airquality.interpolate(method='nearest')

imputations = {'Original': airquality, \
               'Forward Fill': ffill_imputed, \
               'Backward Fill': bfill_imputed, \
               'Linear Interpolation': interpolation_linear, \
               'Quadratic Interpolation': interpolation_quadratic, \
               'Nearest Interpolation': interpolation_nearest}

for ax, df_key in zip(axes.flatten(), imputations):
    imputations[df_key].Ozone.plot(color='red', marker='o', linestyle='dotted', ax=ax)
    airquality.Ozone.plot(marker='o', ax=ax)
    ax.set_title(label=df_key + '- Ozone', fontdict={"fontsize": 18});
```

#### Example 1. Use `KNN` and `MICE` algorithm from `fancyimpute` Library to impute the missing values.

```python
diabetes = pd.read_csv("pima-indians-diabetes data.csv")
diabetes.head()
```

___Use KNN algorithm to impute the missing values___

```python
# Import KNN from fancyimpute
from fancyimpute import KNN

# Copy diabetes to diabetes_knn_imputed
diabetes_knn_imputed = diabetes.copy(deep=True)

# Initialize KNN
knn_imputer = KNN()

# Impute using fit_tranform on diabetes_knn_imputed
diabetes_knn_imputed.iloc[:, :] = knn_imputer.fit_transform(diabetes_knn_imputed)

diabetes_knn_imputed
```

___Use MICE - Multiple Imputations___

```python
# Import IterativeImputer from fancyimpute
from fancyimpute import IterativeImputer

# Copy diabetes to diabetes_mice_imputed
diabetes_mice_imputed = diabetes.copy(deep=True)

# Initialize IterativeImputer
mice_imputer = IterativeImputer()
# Impute using fit_tranform on diabetes
diabetes_mice_imputed.iloc[:, :] = mice_imputer.fit_transform(diabetes)

diabetes_mice_imputed
```
